```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans , DBSCAN ,AgglomerativeClustering
from sklearn.metrics import silhouette_score ,silhouette_samples
from sklearn.mixture import GaussianMixture

import matplotlib.pyplot as plt
import numpy as np
X,y = make_blobs(n_samples=100, n_features=2 ,centers=5, cluster_std=0.8)

clus = AgglomerativeClustering(n_clusters=5) #KMeans(n_clusters= 5) ,
#DBSCAN(eps=2) , GaussianMixture(n_components=5) , AgglomerativeClustering(n_clusters=10)
y_p = clus.fit_predict(X)

print('silhouette score:',silhouette_score(X,y_p))
#print('silhouette sample:',silhouette_samples(X,y_p))

X_t = np.transpose(X)
print(X_t.shape)
plt.subplot(1,2,1)
plt.scatter(X_t[0],X_t[1])
plt.subplot(1,2,2)
plt.scatter(X_t[0],X_t[1],c=y_p)
```
