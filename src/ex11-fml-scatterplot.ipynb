```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import numpy as np

datasets = load_iris()
X,y = datasets.data , datasets.target
X_t = np.transpose(X)

#print('X_t',X_t)

plt.suptitle("compare each artttibute")

k = 1
for i in range(len(X_t)):
    for j in range(i+1,len(X_t)):
        plt.xlabel(datasets.feature_names[j])
        plt.ylabel(datasets.feature_names[i])
        plt.subplot(3,2,k)
        plt.scatter(X_t[i],X_t[j],c=y)
        k = k+1
plt.show()
```
