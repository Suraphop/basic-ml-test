```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_sample_images

datasets = load_sample_images()

#print("image data :", datasets.images)
print("len :",len(datasets.images))

plt.suptitle("image")
img = datasets.images[1]
plt.imshow(img)
print(img.shape)
```

```python
plt.suptitle("Sample images")

for i in range(len(datasets.images)):
    img = datasets.images[i]
    plt.subplot(1,len(datasets.images),i+1)
    plt.imshow(img)
    print("Images number ",i+1,"--> shape : ",datasets.images[i].shape)
plt.show()
```
