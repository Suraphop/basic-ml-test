```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt


datasets = load_iris()
X, y = datasets.data,datasets.target
X_t = np.transpose(X)

print("X_t",X_t.shape)


for i in range(len(X_t)):
    print('feature names :',datasets.feature_names[i])
    print('-> mean',np.mean(X_t[i]))
    print('-> std',np.std(X_t[i]))
    print('-> max',np.max(X_t[i]))
    print('-> min',np.min(X_t[i]))
    print('-> med',np.median(X_t[i]))

plt.suptitle(datasets.feature_names[1])

plt.boxplot(X_t[1])
plt.show()    
```
