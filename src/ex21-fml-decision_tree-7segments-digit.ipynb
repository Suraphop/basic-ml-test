```python
#load dataset

import pandas as pd
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import classification_report,confusion_matrix


datasets = pd.read_csv('../datasets/led/led-10.csv',sep=',',header=None)
#print(datasets)

X = datasets.values[:,:7]
#print(X.shape)

y = datasets.values[:,7]
#print(y)

clf =DecisionTreeClassifier(criterion= 'gini')
y_p = cross_val_predict(clf,X,y,cv=3)

print('classification report:',classification_report(y,y_p))
print('confuseion Matrix:',confusion_matrix(y,y_p))

clf.fit(X,y)

plt.figure(figsize=(24,10))
plot_tree(clf,fontsize =8)
plt.show()
```
